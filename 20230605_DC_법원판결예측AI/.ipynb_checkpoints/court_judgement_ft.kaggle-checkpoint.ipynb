```python
facts = df_NLP['facts']

# 문장을 각각의 단어들로 토큰화 한 뒤 리스트로 저장
tks = nltk.tokenize.word_tokenize(facts.str.cat(sep=' ')) 

# PorterStemmer 알고리즘은 영단어의 가지치기 역할을 한다. 예) eating = eat
# 가끔 쌩뚱맞은 단어로 변형할 수 있다. 예) caring -> car, care가 car가 돼버린다. 영어는 복잡하다.
# 출처 : https://www.projectpro.io/recipes/use-porter-stemmer
ps = nltk.stem.porter.PorterStemmer()

# WordNetLemmatizer 알고리즘은 영단어를 원형으로 변형시켜주는 역할을 한다. 예) caring -> care
# PorterStemmer 보다는 조금 더 유의미하게 변형시켜준다. 하지만 단어형태를 바꾸진 않는다. 예) are -> are
# 출처 : https://www.machinelearningplus.com/nlp/lemmatization-examples-python/
lem = nltk.stem.wordnet.WordNetLemmatizer()

# spacy는 위의 방법들 보다는 더 단어의 원형태로 변형시켜준다.
# 단 pronouns는 -PRON-으로 나타냄을 볼 수 있다.
# 출처 : https://www.machinelearningplus.com/nlp/lemmatization-examples-python/
nlp = spacy.load('en_core_web_sm', disable=['parser','ner'])

# 불용어 제거
stwrds = nltk.corpus.stopwords.words('english')
```
